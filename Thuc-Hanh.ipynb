### Nhập vào ba biến `diem_toan`, `diem_ly`, `diem_hoa` và in ra điểm trung bình của ba môn học trên.

```python
diem_toan = float(input("Nhap vao diem toan: "))
diem_ly = float(input("Nhap vao diem ly: "))
diem_hoa = float(input("Nhap vao diem hoa: "))
diem_trung_binh = (diem_toan+diem_ly+diem_hoa)/3
print(f"diem trung binh = ({diem_toan} + {diem_ly} + {diem_hoa})/3 = {diem_trung_binh}")
```
